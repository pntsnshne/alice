```python
import pandas as pd
import csv
import re
```

```python
data = pd.read_fwf('icd10cm_codes_2019.txt' ,  sep = ' ', header=None, engine = 'python', names = ['code','diagnosis'])
```

```python
#making sure csv split was correct, should be False
data.diagnosis.isnull().any()
display(data)
```

```python
#split icd-10 codes by decimal point after 3rd digit = standard form
add_dec = lambda x: x[:3]+ '.' + x[3:]
data.code = data.code.apply(add_dec)
```

```python
#set index as updated code column
data.set_index('code',inplace= True)
```

```python
display(data)
```

```python
#save to csv
data.to_csv('ICD_10_codes.csv')
```
